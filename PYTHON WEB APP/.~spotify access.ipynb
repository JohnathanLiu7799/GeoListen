```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
from spotipy import util
import pandas as pd
```

```python
token = util.prompt_for_user_token(redirect_uri = 'http://example.com/callback')
token
```

```python
print('USERNAME: ')
username = input()
scope = 'user-read-currently-playing,user-read-recently-played,user-library-read'
token = util.prompt_for_user_token(username, scope, redirect_uri = 'http://example.com/callback')

sp = spotipy.Spotify(auth=token)
```

```python
results = sp.current_user_recently_played()
```

```python
temp_dict = {}
for idx, track in enumerate(results['items']):
    temp_dict[idx] = {}
    temp_dict[idx]['Track Name'] = track['track']['name']
    temp_dict[idx]['Artist'] = track['track']['artists'][0]['name']
```

```python
recently_played = pd.DataFrame(temp_dict).T
recently_played
```

```python
for idx, track in enumerate(results['items']):
    print(f" {idx} \t| {track['track']['name']} - {track['track']['artists'][0]['name']}")
```
